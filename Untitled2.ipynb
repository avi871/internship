**In this task, we create a decision tree classifier and visualize graphically**

```
from sklearn.datasets import load_iris
import pandas as pd

iris=load_iris()

x=pd.DataFrame(iris.data, columns=iris.feature_names)
print(x.head(5))

y=iris.target
print(y)
```

```
x = x.values
```

We plot the data to visualize it 

```
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap = 'rainbow')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.subplot(2,3,2)
plt.scatter(x[:, 0], x[:, 2], c=y, cmap = 'rainbow')
plt.xlabel('Sepal length')
plt.ylabel('Petal length')

plt.subplot(2,3,3)
plt.scatter(x[:, 0], x[:, 3], c=y, cmap = 'rainbow')
plt.xlabel('Sepal length')
plt.ylabel('Petal width')

plt.subplot(2,3,4)
plt.scatter(x[:,1], x[:, 2], c=y, cmap = 'rainbow')
plt.xlabel('Sepal width')
plt.ylabel('Petal length')

plt.subplot(2,3,5)
plt.scatter(x[:,1], x[:, 3], c=y, cmap = 'rainbow')
plt.xlabel('Sepal width')
plt.ylabel('Petal width')

plt.subplot(2,3,6)
plt.scatter(x[:,2], x[:, 3], c=y, cmap = 'rainbow')
plt.xlabel('Petal length')
plt.ylabel('Petal width')

plt.show()
```

```
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.25)
```

Generate the classifier

```
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

```
y_pred = dtc.predict(x_test)

accuracy = dtc.score(x_train, y_train)

accutest = accuracy_score(y_test, y_pred)

confusionmatrix = confusion_matrix(y_test, y_pred)

print(accuracy) 
print(accutest) 
print(confusionmatrix)
```

visualize the tree classifier

```
from sklearn import tree

plt.figure(figsize=(15,10))

tree.plot_tree(dtc, filled = True)

plt.title('tree_plot')
```

```
import numpy as np
 
plt.figure(figsize=(15,10))

for pairidx, pair in enumerate([[0, 1], [0, 2], [0, 3],
                                [1, 2], [1, 3], [2, 3]]):
    # We only take the two corresponding features
    X = iris.data[:, pair]
    y = iris.target

    # Train
    clf = DecisionTreeClassifier().fit(X, y)

    # Plot the decision boundary
    plt.subplot(2, 3, pairidx + 1)

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1))

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)

    plt.xlabel(iris.feature_names[pair[0]])
    plt.ylabel(iris.feature_names[pair[1]])
    plt.axis("tight")

    # Plot the training points
    for i, color in zip(range(3), 'rbg'):
        idx = np.where(y == i)
        plt.scatter(X[idx, 0], X[idx, 1], c=color, label=iris.target_names[i],
                    cmap=plt.cm.Paired)
        plt.axis("tight")

plt.suptitle("Decision surface of a decision tree using paired features")
plt.legend()
plt.show()
```
