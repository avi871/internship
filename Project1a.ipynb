```python
from imageai.Detection import VideoObjectDetection
import os

execution_path = os.getcwd()

detector = VideoObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join('/content/drive/MyDrive/project' , "yolo.h5"))
detector.loadModel()

video_path = detector.detectObjectsFromVideo(input_file_path=os.path.join( '/content/drive/MyDrive/project', "traffic-mini1.mp4"),
                                output_file_path=os.path.join('/content/drive/MyDrive/project', "traffic_mini_detected_2")
                                , frames_per_second=29, log_progress=True)
print(video_path)
```
