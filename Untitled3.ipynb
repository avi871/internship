 **In this task, we had to predict the optimum number of clusters in the iris dataset**

Importing the dataset from sklearn

```
from sklearn.datasets import load_iris
import pandas as pd

iris=load_iris()

x=pd.DataFrame(iris.data, columns=iris.feature_names)

y=iris.target
```

```
x = x.values
```

plotting the data 

```
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap = 'rainbow')
plt.xlabel('Sepal length')
plt.ylabel('Sepal Width')

plt.subplot(2,3,2)
plt.scatter(x[:, 0], x[:, 2], c=y, cmap = 'rainbow')
plt.xlabel('Sepal length')
plt.ylabel('petal length')

plt.subplot(2,3,3)
plt.scatter(x[:, 0], x[:, 3], c=y, cmap = 'rainbow')
plt.xlabel('Sepal length')
plt.ylabel('petal width')

plt.subplot(2,3,4)
plt.scatter(x[:,1], x[:, 2], c=y, cmap = 'rainbow')
plt.xlabel('Sepal width')
plt.ylabel('Petal length')

plt.subplot(2,3,5)
plt.scatter(x[:,1], x[:, 3], c=y, cmap = 'rainbow')
plt.xlabel('Sepal width')
plt.ylabel('Petal width')

plt.subplot(2,3,6)
plt.scatter(x[:,2], x[:, 3], c=y, cmap = 'rainbow')
plt.xlabel('Petal length')
plt.ylabel('Petal width')

plt.show()
```

Applied **Elbow Method** to determine the number of clusters
# Here we are using the **KMeans algorithm** to predict the optimum number of clusters
# It is a unsupervised machine learning algorithm

```
from sklearn.cluster import KMeans

wcss = []

for i in range(1,10):
    kimeans = KMeans(n_clusters = i, max_iter = 500, n_init = 10)
    kimeans.fit(x)
    wcss.append(kimeans.inertia_)

plt.plot(range(1,10), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

From the above plot it is quite clear that three is the optimum number of cluster as we saw huge improvement in three and after that the curve nearly parallel

```
plt.figure(figsize=(12,3))
kmeans3 = KMeans(n_clusters = 3)
ypred = kmeans3.fit_predict(x)
kmeans3.cluster_centers_

```

Finally we plot the data with their respective cluster centers and visualize the prediction by our algorithm

```
plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap = 'rainbow')
plt.scatter(kmeans3.cluster_centers_[:,0],kmeans3.cluster_centers_[:,1], s=150)

plt.subplot(2,3,2)
plt.scatter(x[:, 0], x[:, 2], c=y, cmap = 'rainbow')
plt.scatter(kmeans3.cluster_centers_[:,0],kmeans3.cluster_centers_[:,2], s=150)

plt.subplot(2,3,3)
plt.scatter(x[:, 0], x[:, 3], c=y, cmap = 'rainbow')
plt.scatter(kmeans3.cluster_centers_[:,0],kmeans3.cluster_centers_[:,3], s=150)

plt.subplot(2,3,4)
plt.scatter(x[:, 1], x[:, 2], c=y, cmap = 'rainbow')
plt.scatter(kmeans3.cluster_centers_[:,1],kmeans3.cluster_centers_[:,2], s=150)

plt.subplot(2,3,5)
plt.scatter(x[:, 1], x[:, 3], c=y, cmap = 'rainbow')
plt.scatter(kmeans3.cluster_centers_[:,1],kmeans3.cluster_centers_[:,3], s=150)

plt.subplot(2,3,6)
plt.scatter(x[:, 2], x[:, 3], c=y, cmap = 'rainbow')
plt.scatter(kmeans3.cluster_centers_[:,2],kmeans3.cluster_centers_[:,3], s=150)


plt.show()
```
